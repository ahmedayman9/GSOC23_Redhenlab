## Reading Data

```python
def data_info(data):
    cols = []
    unique = []
    n_uniques = []
    dtypes = []
    nulls = []
    for col in data.columns:
        cols.append(col)
        dtypes.append(data[col].dtype)
        n_uniques.append(data[col].nunique())
        unique.append(data[col].unique())
        nulls.append(data[col].isna().sum())
        
    return pd.DataFrame({'Col' : cols , 'n_uniques' : n_uniques , 
                         'unique' :unique , 'dtypes' : dtypes , "NULLS" : nulls 
                        })
```

## Data cleaning

```python
data.dropna(inplace=True)
```

```python
data['end_time'] = data['end_time'].astype(float )
```

```python
data['length'] = 0

for i in range(1, len(data['end_time'])):
    time_diff = abs(data['end_time'].iloc[i] - data['end_time'].iloc[i-1])
    
    # Update 'length' column with the calculated time difference.
    data.at[i-1, 'length'] = time_diff
```
